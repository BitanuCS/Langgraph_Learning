```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langgraph.graph import StateGraph, START, END
from dotenv import load_dotenv
from typing import TypedDict
```

```python
load_dotenv()

llm = HuggingFaceEndpoint(
    repo_id='Qwen/Qwen2.5-7B-Instruct',
    task='text-generation'
)

model = ChatHuggingFace(llm=llm)
```

```python
class llm_qa(TypedDict):
    ques : str
    ans : str
```

```python
def question(state: llm_qa) -> llm_qa:
    
    question = state['ques']
    
    prompt = f'answer the following question {question}'
    
    model_answer = model.invoke(prompt).content
    
    state['ans'] = model_answer
    
    return state
```

```python
state = StateGraph(llm_qa)

state.add_node('question', question)

state.add_edge(START, 'question')
state.add_edge('question', END)

workflow = state.compile()
```

```python
initial_state = {'ques': 'What is the temparature of the sun?'}

final_state = workflow.invoke(initial_state)

print(final_state['ans'])
```
